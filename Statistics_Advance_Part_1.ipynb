1. What is a random variable in probability theory
 ->A random variable is a variable whose possible values are determined by chance events. It's a function that assigns a numerical value to each outcome of an experiment.
2. What are the types of random variables
 ->Types of Random Variables are
a. Discrete Random Variables: Take on distinct, separate values (e.g., number of heads in coin tosses).
b. Continuous Random Variables: Can take any value within a given range or interval (e.g., height of a person).
3. What is the difference between discrete and continuous distributions
 ->Discrete vs. Continuous Distributions
- Discrete Distributions: Have distinct, separate values (e.g., binomial, Poisson).
- Continuous Distributions: Have values that can be any real number within a given range (e.g., normal, uniform).
4.  What are probability distribution functions (PDF)
 ->Probability Distribution Functions (PDF)
A PDF describes the probability of each possible value of a random variable. For discrete variables, it's the probability of each distinct value. For continuous variables, it's the probability density of the variable within a given range.
5.  How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)
 ->Cumulative Distribution Functions (CDF) vs. PDF
- PDF: Describes the probability of each specific value.
- CDF: Describes the cumulative probability of all values up to a given point.
6. What is a discrete uniform distribution
 ->A discrete uniform distribution is a distribution where every possible value has an equal probability of occurring.
7. What are the key properties of a Bernoulli distribution
 ->A Bernoulli distribution models a single trial with two possible outcomes (success or failure). Key properties:
- Only two possible outcomes
- Probability of success (p) and failure (q = 1 - p) are constant
8. What is the binomial distribution, and how is it used in probability
 ->A binomial distribution models the number of successes in a fixed number of independent trials, each with a constant probability of success. Used in probability to calculate the likelihood of achieving 'k' successes in 'n' trials.
9.  What is the Poisson distribution and where is it applied)
 ->A Poisson distribution models the number of times an event occurs in a fixed interval of time or space, where the events occur independently and at a constant average rate.
10. What is a continuous uniform distribution)
 ->A continuous uniform distribution is a distribution where every value within a given range has an equal probability of occurring.
11. What are the characteristics of a normal distribution)
  ->A normal distribution, also known as a Gaussian distribution or bell curve, is a continuous distribution where the majority of the data points cluster around the mean, tapering off gradually towards the extremes.
12.  What is the standard normal distribution, and why is it important
  ->A standard normal distribution is a normal distribution with a mean of 0 and a standard deviation of 1. It's used as a reference distribution to compare and standardize other normal distributions.
13. What is the Central Limit Theorem (CLT), and why is it critical in statistics)
 ->The CLT states that, given certain conditions, the distribution of the mean of a large sample of independent and identically distributed random variables will be approximately normally distributed, regardless of the original distribution.
14. How does the Central Limit Theorem relate to the normal distribution
 ->Relationship between CLT and Normal Distribution
The CLT explains why many natural phenomena and statistical distributions tend to follow a normal distribution, as the mean of a large sample will converge to a normal distribution.
15. What is the application of Z statistics in hypothesis testing)
 ->Z statistics measure the number of standard deviations between the observed value and the expected value under the null hypothesis. Used to determine whether the observed difference is statistically significant.
16. How do you calculate a Z-score, and what does it represent)
 ->Z = (X - μ) / σ, where X is the observed value, μ is the population mean, and σ is the population standard deviation.
17. What are point estimates and interval estimates in statistics)
  ->Point Estimates and Interval Estimates
- Point Estimates: A single value used to estimate a population parameter.
- Interval Estimates: A range of values used to estimate a population parameter, providing a margin of error.
18. What is the significance of confidence intervals in statistical analysis
 ->A confidence interval provides a range of values within which a population parameter is likely to lie, with a certain level of confidence (e.g., 95%).
19. What is the relationship between a Z-score and a confidence interval)
 ->Relationship between Z-score and Confidence Interval
A Z-score can be used to construct a confidence interval by determining the number of standard deviations from the mean that correspond to a given confidence level.
20. How are Z-scores used to compare different distributions)
 ->Z-scores can be used to compare values from different distributions by standardizing the values to a common scale.
21. What are the assumptions for applying the Central Limit Theorem)
 ->Assumptions for Applying CLT
a. Independence: Observations should be independent of each other.            b. Identical Distribution: Observations should come from the same distribution.
c. Large Sample Size: The sample size should be sufficiently large.
22.  What is the concept of expected value in a probability distribution)
  ->The expected value of a random variable is the long-run average value that the variable is expected to take on when the experiment is repeated many times.
23. How does a probability distribution relate to the expected outcome of a random variable
 ->Relationship between Probability Distribution and Expected Outcome
The expected value of a random variable is a weighted average of the possible values, where the weights are the probabilities of each value occurring. This provides a measure of the central tendency of the distribution.

```python
#2 Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)
import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_distribution(min_val, max_val):
    values = np.arange(min_val, max_val + 1)
    probabilities = np.ones(len(values)) / len(values)
    return values, probabilities

values, probabilities = discrete_uniform_distribution(1, 10)
plt.plot(values, probabilities, 'o-')
plt.title("Discrete Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.show()
```

```python
#3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
import numpy as np

def bernoulli_pdf(p, x):
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0

p = 0.5
x_values = np.array([0, 1])
pdf_values = np.array([bernoulli_pdf(p, x) for x in x_values])
print("Bernoulli Distribution PDF:", pdf_values)
```

```python
#4 Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram
import numpy as np
import matplotlib.pyplot as plt

def binomial_distribution_simulation(n, p, num_trials):
    trials = np.random.binomial(n, p, num_trials)
    return trials

n = 10
p = 0.5
num_trials = 1000
trials = binomial_distribution_simulation(n, p, num_trials)
plt.hist(trials, bins=np.arange(0, n + 2) - 0.5, align='left', rwidth=0.8)
plt.title("Binomial Distribution Simulation")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
#5 Create a Poisson distribution and visualize it using Python
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt
import math


def poisson_distribution_visualization(lam, max_k):
    k_values = np.arange(0, max_k + 1)

    probabilities = np.exp(-lam)  * (lam ** k_values) / [math.factorial(k) for k in k_values]
    plt.plot(k_values, probabilities, 'o-')
    plt.title("Poisson Distribution")
    plt.xlabel("Number of Events")
    plt.ylabel("Probability")
    plt.show()

lam = 5
max_k = 15
poisson_distribution_visualization(lam, max_k)


```

```python
#6 Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution>
import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_distribution_cdf(min_val, max_val):
    values = np.arange(min_val, max_val + 1)
    cdf_values = np.arange(1, len(values) + 1) / len(values)
    return values, cdf_values

values, cdf_values = discrete_uniform_distribution_cdf(1, 10)
plt.plot(values, cdf_values, 'o-')
plt.title("Discrete Uniform Distribution CDF")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.show()
```

```python
# 7 Generate a continuous uniform distribution using NumPy and visualize it>
import numpy as np
import matplotlib.pyplot as plt

def continuous_uniform_distribution(min_val, max_val, num_points):
    values = np.linspace(min_val, max_val, num_points)
    probabilities = np.ones(num_points) / (max_val - min_val)
    return values, probabilities

values, probabilities = continuous_uniform_distribution(0, 1, 100)
plt.plot(values, probabilities)
plt.title("Continuous Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.show()

```

```python
# 10  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

np.random.seed(0)
num_samples = 1000
sample_size = 30

exponential_rvs = expon.rvs(scale=1, size=(num_samples, sample_size))
sample_means = np.mean(exponential_rvs, axis=1)

plt.hist(sample_means, bins=30, density=True)
x = np.linspace(sample_means.min(), sample_means.max(), 100)
y = np.exp(-((x - sample_means.mean()) / sample_means.std())**2 / 2) / (sample_means.std() * np.sqrt(2 * np.pi))
plt.plot(x, y, 'r--')
plt.title("Central Limit Theorem (CLT) Demonstration")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
#15 Implement hypothesis testing using Z-statistics for a sample dataset5

import numpy as np
from scipy.stats import norm

def hypothesis_testing(sample_mean, population_mean, sample_std, sample_size):
    z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(np.abs(z_score)))
    return z_score, p_value

sample_mean = 10.5
population_mean = 10
sample_std = 2
sample_size = 36

z_score, p_value = hypothesis_testing(sample_mean, population_mean, sample_std, sample_size)
print("Z-Score:", z_score)
print("P-Value:", p_value)
```

```python
#16 Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence_level):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)
    z_score = norm.ppf(1 - (1 - confidence_level) / 2)
    margin_of_error = z_score * (sample_std / np.sqrt(sample_size))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    return lower_bound, upper_bound

data = np.array([23, 11, 19, 24, 17])
confidence_level = 0.95
lower_bound, upper_bound = confidence_interval(data, confidence_level)
print("Confidence Interval: ({}, {})".format(lower_bound, upper_bound))

```

```python
#21 Write a Python program to compare the standard deviations of two datasets and visualize the difference

import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(data1, data2):
    std1 = np.std(data1)
    std2 = np.std(data2)
    print("Standard Deviation of Data1:", std1)
    print("Standard Deviation of Data2:", std2)

    plt.hist(data1, bins=30, density=True, alpha=0.5, label='Data1')
    plt.hist(data2, bins=30, density=True, alpha=0.5, label='Data2')
    plt.legend()
    plt.title("Comparison of Standard Deviations")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.show()

data1 = np.array([23, 11, 19, 24, 17])
data2 = np.array([15, 20, 18, 22, 16])
compare_standard_deviations(data1, data2)
```

```python
#22  Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution5

import numpy as np

def calculate_range_and_iqr(data):
    range_ = np.ptp(data)
    q75, q25 = np.percentile(data, [75, 25])
    iqr = q75 - q25
    return range_, iqr

data = np.random.normal(0, 1, 100)
range_, iqr = calculate_range_and_iqr(data)
print("Range:", range_)
print("Interquartile Range (IQR):", iqr)
```

```python
#23  Implement Z-score normalization on a dataset and visualize its transformation5

import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

data = np.array([23, 11, 19, 24, 17])
z_scores = z_score_normalization(data)

plt.hist(data, bins=30, density=True, alpha=0.5, label='Original Data')
plt.hist(z_scores, bins=30, density=True, alpha=0.5, label='Z-Score Normalized Data')
plt.legend()
plt.title("Z-Score Normalization")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#24 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_and_kurtosis(data):
    skewness = skew(data)
    kurtosis_ = kurtosis(data)
    return skewness, kurtosis_

data = np.random.normal(0, 1, 100)
skewness, kurtosis_ = calculate_skewness_and_kurtosis(data)
print("Skewness:", skewness)
print("Kurtosis:", kurtosis_)
```
